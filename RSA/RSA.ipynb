## RSA Starter 5

**-Solution:**

. First, use [factordb](http://factordb.com/) to find the factors of $n$

. Having had $p$ and $q$, now calculate in order:
$$\begin{align*}\phi  &= \left( {p - 1} \right) \times \left( {q - 1} \right)\\d &= {e^{ - 1}}\bmod \phi \\m &= {c^d}\bmod n\end{align*}$$

**-Code:**

````python
from Crypto.Util.number import inverse
e = 65537
p = 857504083339712752489993810777
q = 1029224947942998075080348647219
n = p * q
phi = (p - 1) * (q - 1)
d = inverse(e, phi)
c = 77578995801157823671636298847186723593814843845525223303932
print(pow(c, d, n))
#13371337
````

## Null or Never

**-Solution:**

. This is an implementation of an attack known as the Franklin-Reiter Related Message Attack, which is a type of RSA attack that exploits the homomorphic property of RSA encryption. The attack involves finding a related message to a given ciphertext that decrypts to a known plaintext.

. There will be attempts to recover the plaintext from a given RSA ciphertext encrypted with a small public exponent e=3. It uses the fact that the plaintext is padded with a fixed-length block of zeros, and tries to recover the original message by brute-forcing the possible length of the padding.

. It generates a list of possible values for the padded message, based on the possible length of the padding. Then, for each possible padded message, it tries to calculate the e-th root of the ciphertext, which should give the original message plus some multiple of n. It checks if the resulting message contains the string "crypto", and if so, prints the message. If none of the padded messages work, it means that the padding length is longer than expected or that there is some other issue with the ciphertext.

## Let's Decrypt

**-Solution:**

. Write $m$ to represent the message and $\sigma$ the signature we got. We don't care about the public key of the server, so we only reference our own public key with $N$ or $e$.

$$\sigma > m \Rightarrow \sigma \equiv m \left({\mod \sigma - m}\right)$$

When we then provide $\sigma$ as the signature, we have a good signature for $m$, since $\sigma\mod(\sigma−m)=\sigma−(\sigma−m)=m$, and thus, we easily obtain the parameters to use for our own public key. We wouldn't be able to decrypt anything or sign anything else without factoring $N$ (apart from the fact that $e$ is indeed $11$, so everything is trivial), but then again, we don't have to.

**-Code**

````python
from pwn import *
from json import dumps, loads
from gmpy2 import next_prime
from Crypto.Util.number import *
import pkcs1

p = int(next_prime(2^256))
q = int(next_prime(p))
N = p * q

io = remote("socket.cryptohack.org", 13391)

io.recvline()
io.sendline(dumps({"option": "get_signature"}))
sig = int(loads(io.recvline())["signature"], 16)
msg = "I am Mallory own CryptoHack.org"
digest = pkcs1.emsa_pkcs1_v15.encode(msg.encode(), 256)
# sig * sig - N = dig
N = sig - bytes_to_long(digest)
e = 1
io.sendline(dumps({"msg": msg, "N": hex(N), "e": hex(e), "option": "verify"}))
print(loads(io.recvline())
#crypto{dupl1c4t3_s1gn4tur3_k3y_s3l3ct10n}
````

## Blinding light ( remind me to the weeknd a little bit :)) )

**-Solution:**

. Express `admin=True` as a long yields $m=459922107199558918501733$.

. Plugging this into a factorizer yields two prime factors:

. $p1=211578328037$ and $p2=2173767566209$, $m = p1 \times p2$

. Asking the server to sign each of these messages individually will yield `pow(p1, D, N)` and `pow(p2, D, N)`, and it will respond since neither of these individual messages is `admin=True` when decoded.

. Now, you can compute the digital signature of `admin=True` by multiplying the two signatures you got! Since `p1^D * p2^D = (p1 * p2) ^ D mod N`.

**-Code:**

````python
import json

from Crypto.Util.number import bytes_to_long, long_to_bytes

from pwn import *

p1 = 211578328037

p2 = 2173767566209

conn = remote('socket.cryptohack.org', 13376)

data = conn.recvline()

resp = dict()

resp['option'] = 'get_pubkey'

conn.send(json.dumps(resp))

data = json.loads(conn.recvline())

N = bytes_to_long(bytes.fromhex(data['N'][2:]))

resp = dict()

resp['option'] = 'sign'

resp['msg'] = long_to_bytes(p1).hex()

conn.send(json.dumps(resp))

data = json.loads(conn.recvline())

s1 = bytes_to_long(bytes.fromhex(data['signature'][2:]))

resp = dict()

resp['option'] = 'sign'

resp['msg'] = long_to_bytes(p2).hex()

conn.send(json.dumps(resp))

data = json.loads(conn.recvline())

s2 = bytes_to_long(bytes.fromhex(data['signature'][2:]))

signature = (s1 * s2) % N

resp = dict()

resp['option'] = 'verify'

resp['msg'] = b'admin=True'.hex()

resp['signature'] = long_to_bytes(signature).hex()

conn.send(json.dumps(resp))

data = json.loads(conn.recvline())

print(data['response'])
#crypto{m4ll34b1l17y_c4n_b3_d4n63r0u5}
````

## Vote 4 Pedro

**-Solution:**

. First, we notice that Pedro isn't particularly strict with the format of the padding: anything before the last null byte gets discarded. This means we can add any multiple of $2^{120}$ to the message, and it will still be valid. So the message is of the form $a \times 2^{120} + k$, where $k$ is bytes_to_long(b"VOTE FOR PEDRO") and $a$ is any integer. So if the signature is $x$, then we want $x^3 \equiv k (\mod 2^{120})$ This cube root is quite easy to take with Sage: x = mod(k, 2^120).nth_root(3). Now, this works as the signature because the public modulus $N$ is way bigger than $2^{120}$. $x^3$ is only around 360 bits long, while the modulus is 2048 bits. ## Let's Decrypt Again

**-Solution:**

. Receives a signature (s) from the server by sending a JSON payload with the option field set to `get_signature`.

. Sends the value of n (which is a smooth RSA modulus) to the server by sending a JSON payload with the option field set to `set_pubkey`. The server sends back a suffix value.

. Constructs three messages by appending the suffix value to some predefined strings, and computes their respective SHA-256 hashes (m1, m2, and m3) after padding with PKCS1-v1.5 encoding.

. Attempts to solve for the private exponent d by computing `e = log(m_i, s) mod n` for each of the three messages.

. Sends each e value and its corresponding message index to the server by sending a JSON payload with the option field set to claim.

. Receives a secret value from the server for each claim.

. Computes the flag by XORing the three secret values together.
